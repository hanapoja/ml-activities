```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing data set

```python
df = pd.read_csv('housing_price.csv')
df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["Size (sq ft)"], df["Price ($1000)"], color='blue', label='Data points')
plt.plot(df["Size (sq ft)"], df["Price ($1000)"], color='red', label='Trend line')

# Adding titles and labels
plt.title("House Prices vs. Size", fontsize=14)
plt.xlabel("Size (sq ft)", fontsize=12)
plt.ylabel("Price ($1000)", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
